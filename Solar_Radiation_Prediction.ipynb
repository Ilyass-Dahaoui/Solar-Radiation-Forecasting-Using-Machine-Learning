```python
def api(start_date, end_date, latitude, longitude):
    """api call function.

    Keyword arguments:
    time -- desired date in "%Y-%m-%d format
    latitude -- gps coordinate
    longitude -- gps coordinate
    """

    date_string_1 = start_date + "T" + "00:00"
    date_string_2 = end_date + "T" + "00:00"

    date_object1 = datetime.strptime(date_string_1, "%Y-%m-%dT%H:%M")
    date_object2 = datetime.strptime(date_string_2, "%Y-%m-%dT%H:%M")

    date_only1 = date_object1.date()
    date_only2 = date_object2.date()
    date_string_1 = str(date_only1)
    date_string_2 = str(date_only2)
    # import dataset from API
    x = [(latitude, longitude)]

    li = []
    for i in x:
        params = {
            "latitude": i[0],
            "longitude": i[1],
            "start_date": date_string_1,
            "end_date": date_string_2,
            "timezone": "auto",
            "temperature_unit": "fahrenheit",  #  units
            "windspeed_unit": "mph",
            "precipitation_unit	": "inch",
            "hourly": {
                "precipitation",
                "snowfall",
                "temperature_2m",
                "relativehumidity_2m",
                "surface_pressure",
                "windspeed_10m",
                "winddirection_10m",
                "windgusts_10m",
                "cloudcover",
            },
            "daily": {"sunrise", "sunset"},
        }

        response = requests.get(
            "https://archive-api.open-meteo.com/v1/era5", params=params
        )
        res = response.json()


    return  res
```

```python
import requests
import pandas as pd
from datetime import datetime

def api(start_date, end_date, latitude, longitude):
    """Fonction d'appel à l'API Open-Meteo.

    Arguments:
    start_date -- date de début au format "AAAA-MM-JJ"
    end_date -- date de fin au format "AAAA-MM-JJ"
    latitude -- latitude GPS
    longitude -- longitude GPS
    """
    # Formater les dates
    date_string_1 = start_date + "T00:00"
    date_string_2 = end_date + "T00:00"

    # Paramètres de l'API
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": date_string_1[:10],
        "end_date": date_string_2[:10],
        "timezone": "auto",
        "temperature_unit": "fahrenheit",
        "windspeed_unit": "mph",
        "precipitation_unit": "inch",
        "hourly": [
            "precipitation",
            "snowfall",
            "temperature_2m",
            "relativehumidity_2m",
            "surface_pressure",
            "windspeed_10m",
            "winddirection_10m",
            "windgusts_10m",
            "cloudcover"
        ],
        "daily": ["sunrise", "sunset"],
    }

    # Requête à l'API
    response = requests.get("https://archive-api.open-meteo.com/v1/era5", params=params)

    # Vérifier que la requête est réussie
    if response.status_code != 200:
        print(f"Erreur lors de la requête API pour ({latitude}, {longitude}): {response.status_code}")
        return None, None

    # Charger la réponse JSON
    res = response.json()

    # Vérification des clés "daily" et "hourly" dans les données de réponse
    if "daily" not in res or "hourly" not in res:
        print(f"Données 'daily' ou 'hourly' manquantes pour ({latitude}, {longitude})")
        return None, None

    # Création des DataFrames à partir des données
    df_daily = pd.DataFrame.from_dict(res["daily"], orient="index").T
    df_daily = df_daily.loc[df_daily.index.repeat(24)].reset_index(drop=True)

    df_hourly = pd.DataFrame.from_dict(res["hourly"], orient="index").T
    df_hourly = df_hourly.assign(
        elevation=res.get("elevation", None),
        latitude=res.get("latitude", None),
        longitude=res.get("longitude", None),
        timezone=res.get("timezone_abbreviation", None)
    )
    df_hourly["sunrise"] = df_daily["sunrise"]
    df_hourly["sunset"] = df_daily["sunset"]

    return df_hourly, res

# Traitement final
frames = []
grouped_conc = merged_df.groupby("id")  # Assurez-vous que merged_df est défini dans votre environnement

for id, group in grouped_conc:
    start_time = str(group["date"].iloc[0].date())
    end_time = str(group["date"].iloc[len(group) - 1].date())
    lat = group["latitude"].iloc[0]
    lon = group["longitude"].iloc[0]

    # Appel de l'API
    frame, res = api(start_time, end_time, lat, lon)

    # Vérifiez que les données ont bien été récupérées
    if frame is not None:
        frame["id"] = id  # Ajout de la colonne id
        frames.append(frame)
    else:
        print(f"Les données pour la station {id} n'ont pas été ajoutées (données manquantes).")

# Concaténation des résultats finaux
if frames:
    final_df = pd.concat(frames, axis=0, ignore_index=True)
    print("Données finales concaténées :", final_df)
else:
    print("Aucune donnée valide n'a été récupérée.")
```
